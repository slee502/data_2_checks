```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
import numpy as np
```

```python
# Pull in data from an API. 
url = "http://universities.hipolabs.com/search?country=United+States"
r = requests.get(url)
```

```python
data = r.json()
df = pd.DataFrame(data)
df.head(20)
```

```python
df.tail(20)
```

```python
# Find and print TWO descriptive statistics about your data.(1)
df.describe()
```

```python
# Find and print TWO descriptive statistics about your data. (2)
df['state-province'].replace('None', np.nan)
null_count = df['state-province'].isnull().sum()
print("Number of Null Values in the state-province column:", null_count)
```

```python
# Write a query in Pandas to select a particular set of your data. 
subset = df.query("`state-province`.notnull()")
print(subset)
```

```python
# Select and print the SECOND AND THIRD columns of your data frame.
subset_columns = subset[['alpha_two_code', 'name']]
print(subset_columns)
```

```python
# Select and print the FIRST 4 rows of you data frame.
subset_rows = subset.head(4)
print(subset_rows)
```
