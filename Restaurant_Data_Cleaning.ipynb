# Fix two issues with the data set

## 1. Remove columns with only NaN values

```python
#Drop the PlaceName, Address2, and Intersection columns
rst_df = rst_df.dropna(axis=1, how='all')
rst_df.head(10)
```

## 2. Remove rows with null values

```python
#check how many null values
print("Null values before dropping: ", rst_df.isnull().sum().sum())
```

```python
#remove rows with null values
rst_df = rst_df.dropna()
```

```python
#check if removing null values worked by checking how many null values 
print("Null values after dropping: ", rst_df.isnull().sum().sum())
```
